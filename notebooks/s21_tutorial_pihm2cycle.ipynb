```python
start = time.time()
with open("/Users/rook/Desktop/pg.infil.txt", "r") as f:
    rows = []
    for line in f:
        rows.append(line.split("\t"))
#     rows = [row for row in csv.reader(f, delimiter="\t")]

results = []
for i, row in enumerate(rows):
    for j in range(1, len(row)):
        results.append({
            '@type': 'Record',
            '@id': f"{i}:{j}",
            'data': row[j],
            'value': row[0]
        })

with open('/tmp/output.3.json', 'w') as f:
    ujson.dump(results, f)
end = time.time()
print("exec: %.5f seconds" % (end - start))
```
